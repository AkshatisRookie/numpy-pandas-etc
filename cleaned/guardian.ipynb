```python
import pandas as pd
import numpy as np
data2 = pd.read_csv("..//raw//guardian_headline.csv",header=None) #relative path represented as ..
data2.columns=['Date3','guardian(headlines)']
print(data2.columns)
```

```python
data2=data2.drop(index=[0])
data2
```

```python
data2['Date3'] = pd.to_datetime(data2['Date3'], format='%Y-%m-%d',errors='coerce')
data2['guardian(headlines)']=pd.to_numeric(data2['guardian(headlines)'],errors='coerce') 
data2=data2.dropna(subset=['Date3'])  #drop rows with invalid data
data2
```

```python
data2.set_index('Date3',inplace=True) 
quarterly2=data2.resample('ME').mean() 
quarterly2.reset_index(inplace=True)  
print(quarterly2)
```

```python
quarterly2.to_csv("guardian.csv", index=False)
```
