```python
import pandas as pd
import numpy as np
data1 = pd.read_csv("..//raw//BAA_bond_yield.csv",header=None) #relative path represented as ..
data1.columns=['Date3','BAA-Bond']
print(data1.columns)
```

```python
#removed the uncessary 0th index row
data1=data1.drop(index=[0])
data1
```

```python
data1.describe() #just to check
#col=data.get('Date',None)
#data2= data[['AAA-Bond']].apply(pd.to_numeric, errors='coerce')
data1
```

```python
data1['Date3'] = pd.to_datetime(data1['Date3'], format='%Y-%m-%d',errors='coerce')
data1['BAA-Bond']=pd.to_numeric(data1['BAA-Bond'],errors='coerce') # since AAA-Bond was object type so converted to numeric data type
#above method checks for the format and then if there is any invalid marks as "Nat=not a time"
data1=data1.dropna(subset=['Date3'])  #drop rows with invalid data
data1
```

```python
# setting date column as index for resampling it for quaterly basis
data1.set_index('Date3',inplace=True) #now setting date column back from index after getting quaterly values
quarterly1=data1.resample('QE').mean() #this gives the mean quarterly data
quarterly1.reset_index(inplace=True)  #puts date name back to its place
print(quarterly1) #data-converted to quarterly
```

```python
quarterly1.to_csv("BAA-Bond_quarterly.csv", index=False)
```
