```python
import pandas as pd
import numpy as np
data2= pd.read_csv("..//raw//BAA_bond_yield.csv",header=None) #relative path represented as ..
data2.columns=['Date3','BAA-Bond']
print(data2.columns)
```

```python
#removed the uncessary 0th index row
data2=data2.drop(index=[0])
data2
```

```python
data2.describe() #just to check
#col=data.get('Date',None)
#data2= data[['AAA-Bond']].apply(pd.to_numeric, errors='coerce')
data2
```

```python
data2['Date3'] = pd.to_datetime(data2['Date3'])
data2=data2.set_index('Date3')
date_range = pd.date_range(start='1990-01-01', end='2024-12-31', freq='D')
data2new = data2.reindex(date_range)
data2interpolate = data2new.copy()
data2interpolate['BAA-Bond'] = data2interpolate['BAA-Bond'].interpolate(method='time', limit_direction='both').fillna(method='ffill').fillna(method='bfill')
data2interpolate.reset_index(inplace=True)  
print(data2interpolate)
```

```python
data2interpolate.to_csv("BAA-Bond_quarterly.csv", index=False)
```
