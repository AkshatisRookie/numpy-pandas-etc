```python
import pandas as pd
import numpy as np
data3 = pd.read_csv("..//raw//DGS30(30 years).csv",header=None) #relative path represented as ..
data3.columns=['Date3','DSG(30 year)']
print(data3.columns)
```

```python
#removed the uncessary 0th index row
data3=data3.drop(index=[0])
data3
```

```python
data3['Date3'] = pd.to_datetime(data3['Date3'], format='%Y-%m-%d',errors='coerce')
data3['DSG(30 year)']=pd.to_numeric(data3['DSG(30 year)'],errors='coerce') 
data3=data3.dropna(subset=['Date3'])  #drop rows with invalid data
data3 = data3.set_index('Date3')  #setting date3 as index
data3['DSG(30 year)'] = data3['DSG(30 year)'].interpolate(method='time', limit_direction='both')  #method time is used coz we are doing time based interpolation
data3
```

```python

quarterly3=data3.resample('ME').mean() 
quarterly3.reset_index(inplace=True)  
print(quarterly3)
```

```python
quarterly3.to_csv("DSG(30 year).csv", index=False)
```
